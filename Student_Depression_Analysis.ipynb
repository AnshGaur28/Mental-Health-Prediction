```python
#  Encoding our data explicitly
for i  in range (len(anxiety_depression_data.index)) :
    
    if(anxiety_depression_data['Sleep_Quality'][i]=='Average'):
         anxiety_depression_data['Sleep_Quality'][i]= 1
    elif (anxiety_depression_data['Sleep_Quality'][i]=='Poor'):
        anxiety_depression_data['Sleep_Quality'][i]= 0
    else :
        anxiety_depression_data['Sleep_Quality'][i]= 2
        
        
    if(anxiety_depression_data['Physical_Activity'][i]=='Moderate'):
        anxiety_depression_data['Physical_Activity'][i] = 1
    elif(anxiety_depression_data['Physical_Activity'][i]=='Low'):
        anxiety_depression_data['Physical_Activity'][i] =  0
    else :
        anxiety_depression_data['Physical_Activity'][i] = 2
        
    if(anxiety_depression_data['Diet_Quality'][i]=='Poor'):
        anxiety_depression_data['Diet_Quality'][i] = 0 
    elif(anxiety_depression_data['Diet_Quality'][i]=='Average'):
        anxiety_depression_data['Diet_Quality'][i] = 1 
    else:
        anxiety_depression_data['Diet_Quality'][i] = 2
        
    if(anxiety_depression_data['Extracurricular_Involvement'][i]=='Low'):
        anxiety_depression_data['Extracurricular_Involvement'][i] = 0 
    elif(anxiety_depression_data['Extracurricular_Involvement'][i]=='Moderate'):
        anxiety_depression_data['Extracurricular_Involvement'][i] = 1 
    else:
        anxiety_depression_data['Extracurricular_Involvement'][i] = 2
        
    if(anxiety_depression_data['Social_Support'][i]=='Low'):
        anxiety_depression_data['Social_Support'][i] = 0 
    elif(anxiety_depression_data['Social_Support'][i]=='Moderate'):
        anxiety_depression_data['Social_Support'][i] = 1 
    else:
        anxiety_depression_data['Social_Support'][i] = 2
        
    if(anxiety_depression_data['Relationship_Status'][i]=='Single'):
        anxiety_depression_data['Relationship_Status'][i] = 0 
    elif(anxiety_depression_data['Relationship_Status'][i]=='In a Relationship'):
        anxiety_depression_data['Relationship_Status'][i] = 1 
    else:
        anxiety_depression_data['Relationship_Status'][i] = 2
    
    
    if(anxiety_depression_data['Counseling_Service_Use'][i]=='Never'):
        anxiety_depression_data['Counseling_Service_Use'][i] = 0 
    elif(anxiety_depression_data['Counseling_Service_Use'][i]=='Occasionally'):
        anxiety_depression_data['Counseling_Service_Use'][i] = 1 
    else:
        anxiety_depression_data['Counseling_Service_Use'][i] = 2
        
        
#     if(anxiety_depression_data['Substace_Use'][i]=='Never'):
#         anxiety_depression_data['Substace_Use'][i] = 0 
#     elif(anxiety_depression_data['Substace_Use'][i]=='Occasionally'):
#         anxiety_depression_data['Substace_Use'][i] = 1 
#     else:
#         anxiety_depression_data['Substace_Use'][i] = 2
```

```python
anxiety_depression_data.dropna(inplace = True)
anxiety_depression_data.reset_index(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
anxiety_depression_data['Family_History'] = LE.fit_transform(anxiety_depression_data['Family_History'])
anxiety_depression_data['Chronic_Illness'] = LE.fit_transform(anxiety_depression_data['Chronic_Illness'])
anxiety_depression_data['Substance_Use'] = LE.fit_transform(anxiety_depression_data['Substance_Use'])
# anxiety_depression_data['Gender'] = LE.fit_transform(anxiety_depression_data['Gender']) 
```

```python
anxiety_depression_data
```

```python
X = anxiety_depression_data.drop(['Depression_Score'] , axis=1)
y = anxiety_depression_data['Depression_Score']
```

```python
X.drop(['index'] , axis=1)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2 , random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
reg = DecisionTreeClassifier().fit(X_train , y_train)
```

```python
y_pred = np.round(reg.predict(X_test))
```

```python
from sklearn.metrics import r2_score , accuracy_score
```

```python
r2_score(y_test , y_pred)
```

```python
accuracy_score(y_test , y_pred)
```
